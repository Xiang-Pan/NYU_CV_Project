```python
data = pd.read_csv('./training.csv')
import numpy as np
import matplotlib.pyplot as plt  
x1=[i for i in range(100)]
y1=data['KITTI - train/mIOU_epoch'][0:100]
x2=[i for i in range(100)]
y2=data['streetview-low_kitti - train/mIOU_epoch'][0:100]
x3=[i for i in range(100)]
y3=data['streetview-high_kitti - train/mIOU_epoch'][0:100]
x4=[i for i in range(100)]
y4=data['cameraview_low_kitti - train/mIOU_epoch'][0:100]
x5=[i for i in range(100)]
y5=data['cameraview_high_kitti - train/mIOU_epoch'][0:100]
x6=[i for i in range(100)]
y6=data['complex_kitti - train/mIOU_epoch'][0:100]
x7=[i for i in range(100)]
y7=data['citiscape_kitti - train/mIOU_epoch'][0:100]


l1=plt.plot(x1,y1,color='magenta',label='kitti (baseline)')
l2=plt.plot(x2,y2,color="r",label='streetview-low->kitti')
l3=plt.plot(x3,y3,color="k",label='streetview-high->kitti')
l4=plt.plot(x4,y4,color="c",label='cameraview-low->kitti')
l5=plt.plot(x5,y5,color="tan",label='cameraview-high->kitti')
l6=plt.plot(x6,y6,color="y",label='complex->kitti')
l7=plt.plot(x7,y7,color="g",label='citiscape->kitti')
plt.title('Training mIoU')
plt.xlabel('Training Epoch', fontsize=12)
plt.ylabel('mIoU', fontsize=12)
plt.legend()
# plt.show()
plt.savefig("./training_miou.png")
```
